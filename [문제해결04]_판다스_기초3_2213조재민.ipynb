1. 결측치 처리 (isnull, fillna, loc 등)
전체 결측치가 존재하는 열의 개수를 출력하세요.

원산지 컬럼의 결측값을 "케냐"로 채우세요.

이벤트가 컬럼의 결측값을 3000으로 채우세요.

카푸치노의 원산지를 "베트남"으로 수정하세요. (loc 사용)

칼로리가 100 이상인 메뉴에 대해 이벤트가를 2000으로 변경하세요.


```python
df = pd.read_csv("cafe_review.csv")
df['원산지'].fillna('케냐',inplace=True)
df['이벤트가'].fillna(3000, inplace=True)
df.loc[df['메뉴'] == '카푸치노', '원산지'] = '베트남'
cond = df['칼로리'] >= 100
df.loc[cond, '이벤트가'] = 2000
df
```

2. 문자열 처리 (replace, split, contains, str.len 등)
'아메리카노'를 '룽고', '녹차'를 '그린티'로 바꾸세요.

변경한 이름을 다시 원래대로 되돌리세요.

메뉴 이름 중 '카페'가 포함된 메뉴는 어떤 것인지 Boolean Series로 확인하세요.

메뉴 이름의 글자 수를 나타내는 메뉴길이 컬럼을 추가하세요.

글자 수가 4자 이상인 메뉴만 필터링해서 출력하세요.

```python
df['메뉴'].str.replace("아메리카노", '룽고')
df['메뉴'].str.replace("녹차", '그린티')
df['메뉴'].str.contains("카페")
df['메뉴길이'] = df['메뉴'].str.len()
df[df['메뉴길이'] >= 4]
```

3. 내장 함수 및 통계 함수 활용
가격의 최댓값, 최솟값, 평균값, 중앙값을 각각 출력하세요.

가격의 25% 분위수 이하인 메뉴만 출력하세요.

이벤트가 컬럼의 최빈값을 구하세요.

칼로리가 100 이상이면 "No", 아니면 "Yes"를 나타내는 먹어도 될까요 컬럼을 추가하세요. (apply)

가격 상위 3개 메뉴의 이름과 가격을 출력하세요.

```python
print("최대값:", df['가격'].max())
print("최소값:", df['가격'].min())
print("평균값:", df['가격'].mean())
print("중앙값:", df['가격'].median())

quantile_25 = df['가격'].quantile(0.25)
print(df[df['가격'] <= quantile_25])
print(df['이벤트가'].mode()[0])

def f(x):
    if x >= 100:
        return "No"
    else:
        return "Yes"

df['먹어도 될까요'] = df['칼로리'].apply(f)
print(df)

print(df.nlargest(3, '가격')[['메뉴', '가격']])
```

4. 프로젝트 주제: 이벤트가와 실제 가격 분석을 통한 마케팅 타겟 추출
목표: 이벤트 가격이 제공된 메뉴 중 실제 할인률이 높은 상품을 추출

과제:

이벤트가가 존재하는 메뉴만 필터링

할인율 = (가격 - 이벤트가) / 가격 컬럼 추가

할인율이 30% 이상인 메뉴 추출

해당 메뉴들을 가격순으로 정렬 후 시각화 (bar plot)

```python
import matplotlib.pyplot as plt

df_event = df[df['이벤트가'].notnull()]
df_event['할인율'] = (df_event['가격'] - df_event['이벤트가']) / df_event['가격']
df_discount = df_event[df_event['할인율'] >= 0.3]
df_sorted = df_discount.sort_values(by='가격', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(df_sorted['메뉴'], df_sorted['가격'])
plt.xlabel('menu')
plt.ylabel('price')
plt.show()
```
